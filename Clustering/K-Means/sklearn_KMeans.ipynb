```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values
```

K-Means Clustering

```python
from sklearn.cluster import KMeans
```

Elbow Method: Deciding Optimal Cluster Count <br />
WCSS : sum of all distance between points within centroid from their respective center <br />
Plotting WCSS for different no. of cluster

```python
import matplotlib.pyplot as plt
WCSS = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 69)
    kmeans.fit(X)
    WCSS.append(kmeans.inertia_) 

plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

So we will be using 5 as no. of clusters for K-Means Clustering <br />
k-means++ : selects initial cluster centers for k-mean clustering in a smart way to speed up convergence.

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 69)
y_kmeans = kmeans.fit_predict(X)
```

Ploting the clusters and Centroid

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], c = 'grey', label = 'Cluster1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], c = 'blue', label = 'Cluster2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], c = 'yellow', label = 'Cluster3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], c = 'pink', label = 'Cluster4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], c = 'red', label = 'Cluster5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 80, c = 'cyan', label = 'Centroid')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
