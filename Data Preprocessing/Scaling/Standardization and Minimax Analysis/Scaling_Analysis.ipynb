```python
import pandas as pd
import numpy as np
```

#Importing necessary modules

```python
df = pd.io.parsers.read_csv(
    'https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv',
     header=None,
     usecols=[0,1,2]
    )
```

Reading Data

```python
df.columns=['label', 'A', 'B']

df.head()
```

# Standardization

Standardizing the features so that they are centered around 0 with a standard deviation of 1. It is important if we are comparing measurements that have different units

```python
from sklearn import preprocessing

std_scale = preprocessing.StandardScaler().fit(df[['A','B']])
df_std = std_scale.transform(df[['A','B']])

print('After Standardization mean should be 0 and Standard Deviation is 1')
print ('  Mean\t\tStandard Deviation \nA %f\t%f\nB %f\t%f'
       %(df_std[:,0].mean(),df_std[:,0].std(),df_std[:,1].mean(),df_std[:,1].std()))
```

# Plotting Input and Standard Scale Data

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(df['A'],df['B'],color='green', label='input scale', alpha=.8)
plt.scatter(df_std[:,0],df_std[:,1],color='cyan', label='Standard scale', alpha=.4)
plt.title('A and B feature of Dataset')
plt.xlabel('Feature - A')
plt.ylabel('Feature - B')
plt.legend()
```

# MinMax Scaling
MinValue after Scaling => 0
MaxValue after Scaling => 1

```python
minMax_scale = preprocessing.MinMaxScaler().fit(df[['A','B']])
df_minMax = minMax_scale.transform(df[['A','B']])

print ('  Min Value\tMax Value\nA %f\t%f\nB %f\t%f'
       %(df_minMax[:,0].min(),df_minMax[:,0].max(),df_minMax[:,1].min(),df_minMax[:,1].max()))
```

# Plotting Input and MinMax Scaled Data

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(df['A'],df['B'],color='green', label='input scale', alpha=.8)
plt.scatter(df_minMax[:,0],df_minMax[:,1],color='blue', label='MinMax scale', alpha=.2)
plt.title('A and B feature of Dataset')
plt.xlabel('Feature - A')
plt.ylabel('Feature - B')
plt.legend()
```

# Plotting Input,Standard and MinMax Scale

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(df['A'],df['B'],color='green', label='input scale', alpha=.8)
plt.scatter(df_minMax[:,0],df_minMax[:,1],color='blue', label='MinMax scale', alpha=.2)
plt.scatter(df_std[:,0],df_std[:,1],color='cyan', label='Standard scale', alpha=.4)
plt.title('A and B feature of Dataset')
plt.xlabel('Feature - A')
plt.ylabel('Feature - B')
plt.legend()
```
